```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from reusable import label_Encoder,oneHot_encoder, cleaning, binning, flag, bmi_category

```

```python
health = pd.read_csv('insurance.csv')

df = pd.DataFrame(health)

df = label_Encoder(df)
df = oneHot_encoder(df)
df = cleaning(df)
df = binning(df)
df = flag(df)
df = bmi_category(df)
df.sample(20)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(data=df.drop(columns=['region_northwest','region_southwest','region_southeast','sex','overweight','underweight','children','age']).corr(), annot=True, cmap='coolwarm')
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=df, x=df['charges'],bins = 23)
plt.show()
```

```python
sns.boxplot(data=df, x=df['high_risks'],y=df['charges'])
plt.xticks([0,1],labels=['less_risk','more_risk'])
plt.title("understanding patient risk of increasing bmi")
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(13,8))
sns.scatterplot(data=df, x=df['bmi'],y=df['charges'],ax=axes[0,0])
sns.scatterplot(data=df, x=df['bmi'],y=df['obese'],ax=axes[0,1])
sns.scatterplot(data=df, x=df['charges'],y=df['high_risks'],ax=axes[1,0])
sns.scatterplot(data=df, x=df['charges'],y=df['smoker'],ax=axes[1,1])

plt.show()
```
